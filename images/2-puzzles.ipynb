Consider the moon image shown in the first part.
It's a greyscale image, with every channel having the same values.
If we randomise two of the channels we can still perceive most of the original.

```python
from random import randint

image = load('moon.tiff')
for row in image:
    for pixel in row:
        pixel[R] = randint(0, 255)
        pixel[B] = randint(0, 255)
show(image)
```

To obtain the original image we set the red and blue channels
equal to the untouched green channel.

```python
for row in image:
    for pixel in row:
        pixel[R] = pixel[B] = pixel[G]
show(image)
```

Depending on how the original image is processed, it may be impossible to get it
back. For example, if we halve the green values, which thus become at most 127,
the random red and blue values become dominant: the resulting image is purplish.

```python
image = load('moon.tiff')
for row in image:
    for pixel in row:
        pixel[R] = randint(0, 255)
        pixel[G] = pixel[G] // 2
        pixel[B] = randint(0, 255)
show(image)
```

If we reverse the effect by doubling the green values,
all grey values are even, and so the resulting image differs from the original.
Out of the possible 256 grey levels in the original image,
the resulting image has at most 128,
but the difference is not noticeable to the naked eye.

```python
for row in image:
    for pixel in row:
        pixel[R] = pixel[G] = pixel[B] = pixel[G] * 2
show(image)
```
